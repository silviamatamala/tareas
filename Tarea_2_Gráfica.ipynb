Tarea 2 Ejercicio 02 Hacer un programa que imprima los números del -100 al 200 que sean divisibles entre 6 (con resto 0)
Esto significa que si el numero es 6, 12, 16, etc. entonces debe imprimirlo, ya que la división de esos numeros por 6 da como resto 0, exacto.




```python
# Hacer un programa que imprima los números del -100 al 200 que sean divisibles entre 6 (con resto 0)
numero_inicial = -100
numero_final = 200
lista = range(numero_inicial, numero_final+1)
for i in lista:
  if i % 6 == 0:
    print(i)
```

Pseudocódigo, Sumador y comparador de dos números. Realiza el algoritmo para un sumador de dos números. Si el resultado es menor a 200 se mostrará el mensaje "menor a 200", si el resultado es mayor a 200, pero menor a 250 mostrará el mensaje "mayor a 200", pero si es mayor a 250 mostrará el mensaje "mayor a 250"

```python
def sumador_comparador(num1, num2):
    suma = num1 + num2
    if suma < 200:
        return "menor a 200"
    elif suma < 250:
        return "mayor a 200"
    else:
        return "mayor a 250"

print(sumador_comparador(120, 30))
print(sumador_comparador(180, 40))
print(sumador_comparador(190, 70))
print(sumador_comparador(100, 90))
print(sumador_comparador(210, 30))
```
